```python
import numpy as np

%matplotlib inline
from matplotlib import pyplot as plt
from IPython.core.pylabtools import figsize
```

Coin' flipping awesome!
We first examine the ML case of estimating the probability of heads-tails in a coin flipping example.
Probability of observing a single outcome is:

$$ P(x_i|p) = p^{x_i} (i-p)^{1-x_i} $$

If we increase the number of outcomes to N, we have:
$$ P(X|p) =  \prod_{i=1}^N p^{x_i} (i-p)^{1-x_i} = p^{\sum x_i} (1-p)^{\sum (1-x_i)} = p^{\sum x_i} (1-p)^{N -\sum x_i} $$
$$ L = ln(P(X|p)) = \sum x_i ln(p) + (N - \sum x_i) ln(1-p)\\ \\$$
If we then take a partial derivative of L by p, we get:

$$ \frac{\partial L}{\partial p} = \frac{\sum x_i}{p} - \frac{N - \sum x_i}{1-p} = 0 \\$$

$$ \hat{p}_{ML} = \frac{\sum x_i}{N} $$

```python
# coin flippin' awesome!
import scipy.stats as stats

# we choose Bernoulli for the coin flip
p_true = 0.5

print 'Here are our prior-free ML point estimates:'
for n in range(5):
    N = 10**n
    data = stats.bernoulli.rvs(p_true, size=N)
    print 'N = %5d, p(ML) = ' % N,
    print sum(data) / float(N)
```

Let's now use a prior for our estimate, and somehow incorporate the fact that we believe the coin to be fair. We now maximize our a posteriori probability:

$$ \hat{p}_{MAP} = \mathrm{argmax}_p prob(p|X)  $$

We first choose our a priori belief to be shaped by Beta distribution. It is conventient since Beta distribution takes on values between 0 and 1 and can take on almost any shape:

$$ prob(p) = \frac{1}{B(\alpha, \beta)} p^{\alpha-1}(1-p)^{\beta-1} $$ 

If we take some time to derive the formula, we will reach the final expression for p:

$$ \hat{p}_{MAP} = \frac{\sum x_i + \alpha - 1}{N + \alpha + \beta - 2} $$

But let's first visually inspect what it is that we chose for our prior, for some different values of alpha and beta:

```python
# Plotting the Beta distribution, for some different values:
dist = stats.beta
x = np.linspace(0, 1, 100)

figsize(20, 8)
alphas = [1, 3, 6, 15]
betas = [1, 3, 6, 15]
i = 1
for a in alphas:
    plt.subplot(1, 4, i)
    for b in betas:
        y = dist.pdf(x, a, b)
        plt.plot(x, y, label='a=%d, b=%d' % (a, b))
    i += 1
    plt.ylim(0, 7)
    plt.legend()
```

So, we can see that by changing alpha and beta, we can expand or narrow down the distribution, where lower values of alpha and beta correspond to wider ranges of p value. We can even start from completely uniform distribution, equivalent to the case when both alpha and beta are equal to 1.
Let's say that we are modestly sure about the fact that
$$p_{true} = 0.5$$

We can then set

$$\alpha = 5, \beta=5$$

and compare our ML and MAP estimates:

```python
# coin flippin' awesome!
import scipy.stats as stats

# we choose Bernoulli for the coin flip
p_true = 0.5
alpha = 5
beta = 5

mls = []
maps = []
print 'Here is a comparison between results with and without a prior:\n'
for n in range(5):
    N = 10**n
    data = stats.bernoulli.rvs(p_true, size=N)
    mls.append(sum(data) / float(N))
    maps.append((sum(data) + alpha - 1) / float(N + alpha + beta - 2))
    print 'N = %5d, p(ML) = %.3f, p(MAP) = %.3f' % (N, mls[-1], maps[-1])
    
```

And for the final exercise, we will perform a Bayesian estimate of the posterior distribution in question. Luckily, beta distribution is the conjugate prior of Bernoulli distribution, so we know that if we choose our prior for p to be beta, our posterior will also be beta distribution, but with different parameters.
To check for derivation of the posterior beta, check: https://www.cs.cmu.edu/~10701/lecture/technote2_betabinomial.pdf.

```python
# coin flippin' awesome!
import scipy.stats as stats

# we choose Bernoulli for the coin flip
p_true = 0.5
alpha = 5
beta = 5

mls = []
maps = []
figsize(20, 8)
# First we plot the prior:
plt.figure()
plt.subplot(1, 5, 1)
prior = dist.pdf(x, alpha, beta)
n_samples = [2, 10, 20, 50, 100, 150, 200, 400]

for i, N in enumerate(n_samples, 1):
    plt.subplot(2, 4, i)
    #N = 10**n
    data = stats.bernoulli.rvs(p_true, size=N)
    mls.append(sum(data) / float(N))
    maps.append((sum(data) + alpha - 1) / float(N + alpha + beta - 2))
    new_alpha = alpha + sum(data)
    new_beta = beta + N - sum(data)
    
    plt.plot(x, prior, label='prior')
    plt.fill_between(x, 0, prior, color="#348ABD", alpha=0.4)
    
    y = dist.pdf(x, new_alpha, new_beta)
    plt.plot(x, y, label='Bayes', color='#6C5B7B')
    plt.fill_between(x, 0, y, color="#6C5B7B", alpha=0.4)
    
    plt.vlines(mls[-1], 0, 17, color='#F67280', label='ML', linestyles='--')
    plt.vlines(maps[-1], 0, 17, color='#2F9599', label='MAP', linestyles='--')
    plt.xlim(-0.05, 1.05)
    
    
    plt.title('N=%d' % N)
    plt.legend()
```
