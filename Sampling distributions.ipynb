# Sampling distributions experiments

This is just a byproduct of learning about sampling distributions. It will touch upon topics such as Central Limit Theorem, distributions of sample statistics and the like.

Let's first just try out what we get if we repeatedly sample different distributions for their mean and variance.

```python
import scipy.stats as stats
import numpy as np

%matplotlib inline
from IPython.core.pylabtools import figsize
from matplotlib import pyplot as plt
```

```python
from collections import defaultdict
import math

distr = stats.norm
_mean = 5.
_var = 1.

figsize(17, 11)
x = np.linspace(0, 10, 1e3)
pdf = distr.pdf(x, _mean, _var)
plt.subplot(4, 1, 1)
plt.plot(x, pdf)
plt.title('Underlying distribution: N(%d, %.1f)' % (_mean, _var))

num_samples = 10000
means = defaultdict(list)
for idx, N in enumerate([5, 10, 20, 50, 100, 1000]):
    for i in xrange(num_samples):
        mean = distr.rvs(_mean, _var, size=N).mean()
        means[N].append(mean)
    plt.subplot(4, 2, 3 + idx)
    n, bins, patches = plt.hist(means[N], 1000)
    dx = bins[1] - bins[0]
    plt.hold(False)
    n_new = n / num_samples / dx
    bins_new = [(bins[i] + bins[i+1])/2 for i in range(len(bins)-1)]
    plt.hist(bins_new, 1000, weights=n_new)
    plt.hold(True)
    plt.plot(x, distr.pdf(x, _mean, _var/math.sqrt(N)), 'r')
    #plt.xlim(distr.ppf(1e-3, _mean, _var/N), distr.ppf(1-1e-3, _mean, _var/N))
    plt.xlim(4, 6)
    plt.title('N = %d' % N)
    
    
```

## Central limit theorem

In probability theory, the central limit theorem (CLT) states that, given certain conditions, the arithmetic mean of sufficiently large number of iterates of independent random variables, each with a well-defined expected value and well-defined variance, will be approximately normally distributed, regardless of the underlying distribution.

## Characteristic function

In probability theory and statistics, the characteristic function of any real-valued random variable completely defines its probability distribution. If a random variable admits a probability density function, then the characteristic function is the inverse Fourier transform of the probability density function. Thus it provides the basis of an alternative route to analytical results compared with working directly with probability density functions or cumulative distribution functions.

For a scalar random variable $X$ the characteristic function is defined as the expected value of $e^{itX}$, there $i$ is the imaginary unit, and $t \in R$ is the argument of the characteristic function:
$$
\phi_X(t) = E[e^{itX}] = \int_R e^{itx}dF_X(x) = \int_R e^{itx}f_X(x)dx
$$
For instance, for the normal distribution, with probability distribution function:
$$
f(x) = \frac{1}{\sigma \sqrt{2 \pi}} e^{- \frac{-(x- \mu)^2}{2 \sigma^2}}
$$
the characteristic function is:
$$
\hat{\phi}(t) = \int_{-\infty}^{+ \infty}e^{itx}f(x)dx = e^{i\mu t} e^{-\frac{1}{2} (\sigma t)^2}
$$

```python
import cmath

norm = stats.norm

x = np.linspace(-5, 5, 100)
figsize(16, 5)


_var = 1.
t = np.linspace(-5, 5, 100)
for _mean in [0, 1, 2, 3]:
    plt.subplot(4,3,_mean * 3 + 1)
    plt.plot(x, norm.pdf(x, _mean, _var))
    ch_fun = np.exp(1j * _mean * t) * np.exp(-0.5 * (_var * t)**2)
    plt.subplot(4,3,_mean * 3 + 2)
    plt.plot(t, np.abs(ch_fun))
    plt.subplot(4,3,_mean * 3 + 3)
    plt.plot(t, np.angle(ch_fun))
    
_mean = 1.5
plt.figure()

t = np.linspace(-40, 40, 1000)
for idx, _var in enumerate([0.01, 0.1, 1, 10, 100]):
    plt.subplot(5,3,idx * 3 + 1)
    plt.plot(x, norm.pdf(x, _mean, _var))
    ch_fun = np.exp(1j * _mean * t) * np.exp(-0.5 * (_var * t)**2)
    plt.subplot(5,3,idx * 3 + 2)
    plt.plot(t, np.abs(ch_fun))
    plt.subplot(5,3,idx * 3 + 3)
    plt.plot(t, np.angle(ch_fun))
    
```

### Sufficiency

In statistics, a statistic is sufficient with respect to a statistical model and its associated unknown parameter if "no other statistic that can be calculated from the same sample provides any additional information as to the value of the parameter".[1] In particular, a statistic is sufficient for a family of probability distributions if the sample from which it is calculated gives no additional information than does the statistic, as to which of those probability distributions is that of the population from which the sample was taken.

### Fisher information

In mathematical statistics, the Fisher information (sometimes simply called information) is a way of measuring the amount of information that an observable random variable X carries about an unknown parameter θ of a distribution that models X. Formally, it is the variance of the score, or the expected value of the observed information. In Bayesian statistics, the asymptotic distribution of the posterior mode depends on the Fisher information and not on the prior (according to the Bernstein–von Mises theorem, which was anticipated by Laplace for exponential families).[2] The role of the Fisher information in the asymptotic theory of maximum-likelihood estimation was emphasized by the statistician Ronald Fisher (following some initial results by Francis Ysidro Edgeworth). The Fisher information is also used in the calculation of the Jeffreys prior, which is used in Bayesian statistics.
